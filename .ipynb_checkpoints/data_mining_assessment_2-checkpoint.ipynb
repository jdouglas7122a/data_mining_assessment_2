# Introduction:

Pneumonia is an affliction of the lungs where various microorganisms, bacteria, fungi, and viruses, implant themselves in the lower respiratory tract, leading to infections and other medical complications including death. The condition can be broken down into Community-Acquired Pneumonia (CAP) and 
Hospital Acquired Pneumonia (HAP).

While manifesting the same symptoms, CAP and HAP cases tend to vary in the pathogens that cause the conditions, with Pneumonia being largely independent of these, and more correlated with the immune response of the host[]. It is worth considering that, although immune response is more responsible for the initial contraction of the condition, treatment is far more dependent on the initial pathogen that caused the infection, an example of this is bacterial pneumonia requiring antibiotic intervention, but further use of this medication contributes to the issue of antibiotic resistance in the population, a growing concern throughout medicine.

The condition tends to manifest in those with weakened immune systems, often leading to those already in a weak state being affected. This includes individuals under the age of 5, and those with chronic health conditions[]. Diagnosis of the condition relies heavily on radiological findings[], with early treatment being vital for recovery and to prevent severe damage to the lining of the lungs[]. 

Once hospitalized, a patient's risk factors increase hugely. Not only this, but the financial implications on hospitalized Pneumonia patients puts a huge strain on the NHS, with complications due to comorbidity being common. This emphasises the importance of early diagnosis before hospitalization is required. But with the NHS already under strain, we need to take advantage of as many diagnostic tools as possible, overusing practicians time. 

Thanks to developments in machine learning technology we can use a variety of models, supervised and unsupervised, including the use of deep learning approaches for identifying pneumonia from chest X-rays resulting in an accuracy rate of 93.57% in 2022[]. Although this approach was slow by today's standards, we can see similar methods replicated, achieving a slightly reduced accuracy, but with huge time reductions, meaning that we can quickly diagnose many cases, and advance individuals to early anti-bacterial medication before we start to see the side effects of onset pneumonia. 

In this notebook we will be attempting to build a similar model whose goal is to quickly identify these pneumonia cases, while keeping technical resource spending low, ensuring that the model can be run on any system for quick early diagnosis.

```python
def display_images_in_grid(image_paths, figure_title, image_size=(200, 200), grid_width=6, grid_height=2):

    # Counts the number of images required for the image
    total_images = grid_height * grid_width

    # Removes excess paths from the list
    image_paths = image_paths[:total_images]
    
    # Set up the figure size based on the number of images and the image size
    plt.figure(figsize=(grid_width * image_size[0] / 100, grid_height * image_size[1] / 100))

    # Create a new blank image for the grid
    grid_image = Image.new('RGB', (image_size[0] * grid_width, image_size[1] * grid_height))

    for index, image_path in enumerate(image_paths):
        # Open the image, resize it, and paste it into the grid
        img = Image.open(image_path).resize(image_size)
        x_offset = (index % grid_width) * image_size[0]
        y_offset = (index // grid_width) * image_size[1]
        grid_image.paste(img, (x_offset, y_offset))
    
    # Display the grid
    plt.imshow(grid_image)
    plt.axis('off')  # Turn off axis numbers and ticks
    plt.title(figure_title)
    plt.show()
```

When we are comparing lungs of patients with vs without pneumonia, there are some key visual indicators we are looking for:
- Consolidation: An area of the lung may appear white or grey sur to the filling of the normally air-filled spaces with pus, fluid or other inflammatory material.  
- Air Bonochograms: These are caused by air-filled bronchi (tubes in the lungs) being made visible by surrounding alveoli filled with fluid, Normally these tubes are not visible in xrays  
- Elevated Diaphram: Due to fluid or inflamation, the diaphrahm my appear elevated.  
- Increased Opacity: Affected areas may appear to be more opaque, this may be be because of inflamation or obstruction from fluid.  
- Plenural Effusions: Sometimes fluid can can collect in the space between the ungs and chest wall.

```python

def create_pie_chart(ax, data, title, colors=['#FF3636', '#3FBE20']):
    """
    Create a pie chart with the given data and title in the specified axes.
    :param ax: The axes object to draw the pie chart on.
    :param data: The data for the pie chart.
    :param title: The title of the pie chart.
    :param colors: The colors to use for the pie chart slices.
    """
    wedges, texts, autotexts = ax.pie(data, colors=colors, autopct=lambda pct: f'{int(round(pct*sum(data)/100.0))}',
                                      startangle=140)

    # Remove the labels on the edge
    for text in texts:
        text.set_visible(False)
    
    # Change the font size and color for the count inside the slices
    for autotext in autotexts:
        autotext.set_color('white')
        autotext.set_fontsize(12)
    
    ax.set_title(title)

def plot_data_split(image_dataset):

    # Example data
    data1 = [len(image_dataset['training']['pneumonia']), 
             len(image_dataset['training']['normal'])]  # Data for the first pie chart
    
    data2 = [len(image_dataset['testing']['pneumonia']), 
             len(image_dataset['testing']['normal'])]      # Data for the second pie chart
    
    data3 = [len(image_dataset['validation']['pneumonia']), 
             len(image_dataset['validation']['normal'])]       # Replace with your actual data
    
    fig, axarr = plt.subplots(1, 3, figsize=(15, 5))
    
    # Create each pie chart
    create_pie_chart(axarr[0], data1, 'Training Data')
    create_pie_chart(axarr[1], data2, 'Testing Data')
    create_pie_chart(axarr[2], data3, 'Validation Data')
    
    # Set the figure title
    plt.suptitle('Data Split by Purpose and Pneumonia Pressence', fontsize=16)
    
    # Create a legend
    plt.figlegend(['Pneumonia Negative', 'Pneumonia Positive'], loc='lower center', ncol=2)
    
    # Adjust the layout
    plt.tight_layout(rect=[0, 0.1, 1, 0.95])
    
    # Show the plots
    plt.show()

def plot_full_data(image_dataset):

    data = [0,0]

    data[0] += len(image_dataset['training']['pneumonia'])
    data[1] += len(image_dataset['training']['normal'])

    data[0] += len(image_dataset['testing']['pneumonia'])
    data[1] += len(image_dataset['testing']['normal'])

    data[0] += len(image_dataset['validation']['pneumonia'])
    data[1] += len(image_dataset['validation']['normal'])
    
    
    fig, axarr = plt.subplots(1, 1, figsize=(15, 5))
    
    # Create each pie chart
    create_pie_chart(axarr, data, 'Total Data')

    
    # Set the figure title
    plt.suptitle('Data Split by Purpose and Pneumonia Pressence', fontsize=16)
    
    # Create a legend
    plt.figlegend(['Pneumonia Negative', 'Pneumonia Positive'], loc='lower center', ncol=2)
    
    # Adjust the layout
    plt.tight_layout(rect=[0, 0.1, 1, 0.95])
    
    # Show the plots
    plt.show()
```

```python
ROTATION_RANGE = 20
```

```python
def load_data(tr_path, v_path, te_path):
    datagen = ImageDataGenerator(
                rescale=1./255,
                validation_split=0.2,
                rotation_range=ROTATION_RANGE, 
                fill_mode='nearest')

    train_generator = datagen.flow_from_directory(
                directory=tr_path,
                target_size=(IMG_HEIGHT, IMG_WIDTH),
                batch_size=BATCH_SIZE,
                class_mode='binary',
                subset='training',
                seed=RANDOM_SEED)

    validation_generator = datagen.flow_from_directory(
                directory=v_path,
                target_size=(IMG_HEIGHT, IMG_WIDTH),
                batch_size=BATCH_SIZE,
                class_mode='binary',
                #subset='validation',
                seed=RANDOM_SEED)

    testing_generator = datagen.flow_from_directory(
                directory=te_path,
                target_size=(IMG_HEIGHT, IMG_WIDTH),
                batch_size=BATCH_SIZE,
                class_mode='binary',
                seed=RANDOM_SEED,
                shuffle=False)
    
    return train_generator, validation_generator, testing_generator

train_generator, validation_generator, testing_generator = load_data(TRAIN_PATH, 
                                                                     VAL_PATH, 
                                                                     TEST_PATH)
```

## Model Selection

### CNN's

This is a common approach to image classification problems due to its high support throughout common machine learning packages such as TensorFlow and its proven track record for image and video classification tasks[]. They are known for hierarchal feature learning, meaning they automatically and adaptively learn spatial hierarchies of features from high and low-level patterns. Further, they use the concept of shared weights in convolutional layers, meaning they don't require many inputs when compared to fully connected networks of comparable size[], this hugely decreases the amount of memory required to run them.  

A large benefit for this task is translation invariance, meaning once trained, the model can identify objects/patterns regardless of its orientation, this is due to the pooling layers that disregard the location and scale of features, ensuring that the model resulting is robust. 

Finally, they are excellent at integrating with other neural networks, almost seamlessly[]. Some examples of this are RNN's (recurrent neural networks) for video analysis[] and LSTM's (Long short-term memory networks) for time series data analysis[].

They do have drawbacks, one of the main being computational cost. This is primarily due to the high number of parameters within the model[] and the extensive data needed for training[]. Additionally, the act of completing convolution operations are computationally intensive, with each convolution requiring multiple multiplications and additions for each pixel in the output. When this is done with multiple layers, each with many filters, the cost can add up quickly.

Ease of application/modification in this project will go a long way and allow us to analyze the impact of modifying layers such as adding more pooling and 2d convolution layers. 

### Capsule Networks:

Capsule networks are far more experimental than cnn's and have a much smaller breath of research surrounding them.

Support for this type of neural network is far more abstract due to its experimental nature[], with no quick way to support the capsule layers its built on. In order to support this type of neural network one would need to build a capsule layer from scratch rather than just importing a layer type from tensorflow.

Aside from its use of capsules, one of the largest difference is the dynamic structure of the network its build on[]. While still using layers like cnn's, they have a dynamic routine process that allows them to send their outputs to appropriate parent capsules in the layer above.l The decision on where to route the information is based on prediction agreement between lower level and higher level and lower level capsules. This assists in identifying part-whole relationships[].

This will be a short look into capsule networks due to the overcomplicated nature of them in current application due to their experimental nature, while they hold promising potential for specific reserach purposes when CNN's dont deliver[], they are still under development and do not consistently our perform CNN's[].

### Transfer Learning with Pre-trained models.

Transfer learning is the idea of using pre-trained models in order to take advantage of the reduced training time. There is a variety of support for these models within TensorFlow or Keras within the tf.keras.applications module, some examples are:

- ResNet
- InceptionV3
- Xception
- VGG16 & VGG19
- EfficientNet

Out of these, we're going to look mainly into Resnet, InceptionV3 and EfficientNet.

ResNet: This model is known for its ability to train deep networks due to its utilization of residual connections which reduce the effect of the vanishing gradient problem[]. This leads to robust feature extraction capabilities that can be helpful when trying to detect the subtle features that might indicate the presence of pneumonia. While this can be incredibly powerful, the deeper versions of ResNet might be much too large a model for this task with this dataset. It may incur very high computational costs and risks overfitting due to smaller datasets[], requiring large amounts of data transformation/regularization in the training set[]. 

InceptionV3: This model uses a variety of kernel sizes in each convolution layer to identify information at a variety of scales. This is particularly useful in medical cases such as pneumonia where signs of illness can vary hugely in size. While useful, the difficulty with using this model can come at the architecture point, where feature tuning becomes particularly important due to its multi-scale nature. 

EfficientNet: This set of models upscales CNNs in a structured way that attempts to balance depth, width and resolution. These optimizations can lead to better performance and the reduction of computational cost in training and utilizing the finished model. This model is particularly good when working in environments where computational cost is at the forefront of importance. There is a tradeoff for this, in the complex nature of fine-tuning to make it adapt to tasks like pneumonia detection, and will require a greater understanding of CNNs and the layers that reside inside. 

With these factors in mind we're going to be attempting to utilize a CNN we build from scratch with minimal layers to demonstrate how effective a simple approach can be when using CNN's, and EfficientNet, attempting to get a high accuracy model while incuring small computational costs.

## Building the CNN

For this we will be building a sequental model using the tensorflow.keras.models module, the layers we will be using include:

Conv2D:
- Designed to perform the 2d convolution.
- Applies a series of filters to the input image to generate feature maps that summarise the pressence of features. 
    - A filter is a small matrix of weights that 'convolves' around the input image
    - Each filter is degined to detect a very specific type of feature within the input image at various locations and rotations.
    - Each filter is convolved over the length of the image, at each position where the filter is applied.
    - The scanning-like process is applied for each filter applied to produce a two dimensional activation map for each filter.
    - Uses a 'stride' value to control how the scanning happens over each pixel, a stride value of 1 moves the filter one pixel per step.
    - Padds the edge of the image with 0's, this means that features at the edge of the image can still be identified.
    - After a feature map has been fully produced, ReLU is typically applied (Rectified Linear Unit), this adds non-linearity to the model which helps with the identification of non-linear patterns. 

MaxPooling2D:

- Designed to reduce spatial dimensions of feature maps which decreases the computational cost and minimizes overfitting[].
- Further, helps the model identify features without regard to their location or orientation.
- The pooling operation acts by sliding a window over the input feature map and outputs a maximum value in each 'window'.
- This windowing process is repeated across the whole map.
- The size of this window is defined as a tuple, typically this is (2,2) which reduces the width and height of the map by half.
- Similar to the conv2d later, it uses a stride value and padding.

Flatten:

- Condenses a multi-dimensional input into a 1d array, this is fed into the connection layers that follow.
    
Dense (relu & sigmoid):

- A dense layer connects every input its given to every output layer above.
- Each neuon in a dense layer calculates a weighted sum of its inputs, adds bias then feeds the output to an activation function.
- This layer reduces linearity
- Relu:
    - The relu activation layer outputs its input if the input is positive, otherwise it outputs zero.
    - This layer is widely used in most convolutional networks because it reduces the likelyhood of the vanishing gradient problem[].
- Sigmoid: 
    - The sigmoid activation layer ouputs a value between zero and one based on the following formula: 1/1+e^−x1. This makes it particularly useful for binary classification problems
    - Used to measure certainty of an outcome as a probability score.
 

Dropout:   
- Aims to prevent overfitting
- This is achieved by randomly setting outputs to 0, which prevents overfitting by reducing the effect of co-adaptation.
- This prevention forces the network to adapt by finding new paths, enhancing the stability out outcomes
  
When training, we want to keep the epochs to a minimum, to further keep our computational cost low.

```python
def build_cnn(train_generator, validation_generator):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_WIDTH, IMG_HEIGHT, 3)),
        MaxPooling2D(2, 2),
        Flatten(),
        Dense(512, activation='relu'),
        Dropout(0.5),
        Dense(1, activation='sigmoid')])

    model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])

    model.summary()

    for epoch in range(5): 
        print(f"Epoch {epoch+1}/{5}")
        try:
            model.fit(
                train_generator,
                steps_per_epoch=train_generator.samples // BATCH_SIZE,
                epochs=1,  # Set epochs to 1 since we are controlling it manually
                validation_data=validation_generator,
                validation_steps=validation_generator.samples // BATCH_SIZE)
            
        except StopIteration:
            # Reset the generator by recreating it
            train_generator = datagen.flow_from_directory(
                directory='./chest_xray/train/',
                target_size=(IMG_HEIGHT, IMG_WIDTH),
                batch_size=BATCH_SIZE,
                class_mode='binary',
                subset='training',
                seed=RANDOM_SEED)
            continue

    return model

cnn_model = build_cnn(train_generator, validation_generator)
```

## EfficientNet

For this, we are not only going to use EfficientNet, but embed it in a sequential model that uses similar layers to our base CNN, with a couple of key differences.

EfficientNetB0:

- This model is one of 8 offered varying in size where B0 is the smallest and B7 is the largest.  
- We are going to utilize the pre-trained weights from ImageNet.
- This model works off the concept of compound scaling in place of arbitrary scaling. This is done by using a compound coefficient to uniformly scale the dimensions of the neural network. 
- The model has been pre-trained on ImageNet, a database of images for use in image recognition research projects. The images within this set are classified into over 20000 classes. This is a common dataset to use when approaching object recognition in images. 

In the sequential model, we add one layer we havent used before, GlobalAveragePooling2D(). 
Global Average Pooling:


```python
IMG_SIZE = (IMG_HEIGHT, IMG_WIDTH)

def build_efficientnet(train_generator, validation_generator):
    
    base_model = EfficientNetB0(input_shape=(*IMG_SIZE, 3), include_top=False, weights='imagenet')
    
    # Freeze the weights of the base model before moving onto the further layers in the sequential model.
    base_model.trainable = False

    # Creates the sequential model on top of the EfficientNet Model
    model = Sequential([
        base_model,
        GlobalAveragePooling2D(),
        Dense(512, activation='relu'),
        Dropout(0.5),
        Dense(1, activation='sigmoid')])
    
    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // BATCH_SIZE,
        epochs=30,
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // BATCH_SIZE)

    model.summary()
    
    return model

efficientnet_model = build_efficientnet(train_generator, validation_generator)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix
from tensorflow.keras.preprocessing.image import ImageDataGenerator
    
    # Assuming `cnn_model` is your trained model and the test_generator is properly set up
    # Predict probabilities
    
def draw_conf_matrix(model, test_generator, figure_title):
    # Assuming `cnn_model` is your trained model
    # Make sure test_generator is properly set up with shuffle=False
    test_generator.reset()  # Reset the generator before prediction
    
    # Calculate the correct number of steps to cover all samples
    steps = int(np.ceil(test_generator.samples / test_generator.batch_size))
    
    # Predict probabilities or outputs
    predictions = cnn_model.predict(test_generator, steps=steps)

    
    # Convert probabilities to binary class labels (if binary classification)
    predicted_classes = (predictions > 0.5).astype(int)
    
    # Extract true labels from the generator
    true_classes = test_generator.classes
    
    # Compute the confusion matrix
    cm = confusion_matrix(true_classes, predicted_classes.flatten())
    
    # Display the confusion matrix using Seaborn for better visualization
    import seaborn as sns
    import matplotlib.pyplot as plt
    
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix: {figure_title}')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()

    loss, accuracy = model.evaluate(
    test_generator,
    steps=test_generator.samples // BATCH_SIZE  # Ensures all test data is used
    )

    return loss, accuracy
```

```python
from sklearn.utils.class_weight import compute_class_weight

def reset_generator(e_count):

    datagen = ImageDataGenerator(
        rescale=1./255,
        validation_split=0.2,
        rotation_range=ROTATION_RANGE+(e_count/2), 
        fill_mode='nearest',
        horizontal_flip=True)
    
    # Reset the generator by recreating it
    train_generator = datagen.flow_from_directory(
        directory=TRAIN_PATH,
        target_size=(IMG_HEIGHT, IMG_WIDTH),
        batch_size=BATCH_SIZE,
        class_mode='binary',
        subset='training',
        seed=RANDOM_SEED + e_count)  # iterates the random seed so new data is produced

    return train_generator

def build_cnn_weighted(train_generator, validation_generator):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_WIDTH, IMG_HEIGHT, 3)),
        MaxPooling2D(2, 2),
        Flatten(),
        Dense(512, activation='relu'),
        Dropout(0.5),
        Dense(1, activation='sigmoid')])

    e_count = 0

    classes = train_generator.classes
    class_weights = compute_class_weight(
        'balanced',
        classes=np.unique(classes),
        y=classes
    )
    class_weight_dict = dict(enumerate(class_weights))
    

    model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])

    model.summary()
    
    train_generator = reset_generator(e_count)

    for epoch in range(10): 
        print(f"Epoch {epoch+1}/{5}")
        try:
            model.fit(
                train_generator,
                steps_per_epoch=train_generator.samples // BATCH_SIZE,
                epochs=1,  # Set epochs to 1 since we are controlling it manually
                validation_data=validation_generator,
                validation_steps=validation_generator.samples // BATCH_SIZE,
                class_weight=class_weight_dict
            )
        except StopIteration:
            e_count += 1
            train_generator = reset_generator(e_count)
            continue

    return model

cnn_model_weighted = build_cnn_weighted(train_generator, validation_generator)
```

```python
IMG_SIZE = (IMG_HEIGHT, IMG_WIDTH)



def build_efficientnet_improved(train_generator, validation_generator):
    
    base_model = EfficientNetB0(input_shape=(*IMG_SIZE, 3), include_top=False, weights='imagenet')
    
    # Freeze the weights of the base model before moving onto the further layers in the sequential model.
    base_model.trainable = False

    # Creates the sequential model on top of the EfficientNet Model
    model = Sequential([
        base_model,
        GlobalAveragePooling2D(),
        Dense(512, activation='relu'),
        Dropout(0.5),
        Dense(1, activation='sigmoid')])
    
    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    
    callbacks = [
        ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=10),
        EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True),
        ModelCheckpoint(filepath='best_model.h5', monitor='val_loss', save_best_only=True)]
    
    model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // BATCH_SIZE,
        epochs=30,
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // BATCH_SIZE,
        callbacks=callbacks)

    model.summary()
    
    return model

efficientnet_model_improved = build_efficientnet_improved(train_generator, validation_generator)
```
